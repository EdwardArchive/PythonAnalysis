# XML

```python
*.XML(Extensible Markup Language)

	1)메타언어
		GML -> SGML -> XML
	2)데이터를 위한 언어
	3)DB
	4)데이터의 표준화 : 이기종 시스템간의 정보교환, 웹서비스, 
		유비쿼터스, 사물인터넷 기타 등등...
	
*. XML 문법

	XML(.xml), DTD(.dtd), XML스키마(.xsd), XSL
   
	(1) XML 문서 종류
        1) Well-Formed XML Document(잘 짜여진 문서)
		2) Valid XML Document(유효한 문서)
			
	(2) XML 기본 구조
		Prolog	- 1개
		Element	- 1개
		Misc	- 0개 이상

	(3) Element 문법
		- 작성 규칙
			a) 모든 XML문서는 반드시 단 하나의 루트 엘리먼트를 가진다.
			b) 루트 엘리먼트는 여러 개의 자식 엘리먼트를 가질 수 있고 또 그 자식도 자신의 자식을 가질 수 있다.
			c) 시작태그와 끝태그는 반드시 짝을 이루어야 한다.
			d) "<"는 값으로 사용 불가. ">"는 사용할 수 있으나가급적 사용 금지
			e) "<"와 ">" 다음에 공백문자가 올 수 없으며 반드시 시작태그와 끝태그의 이름이 같아야 한다.
					
		- 종류
			a) 내용을 가지는 엘리먼트
			b) 내용이 없는 엘리먼트

		- 내용에 대한 종류
			a) 문자 데이터
			b) 자식 엘리먼트
			c) 엔티티 또는 문자 참조(&엔티티명;)
			d) CDATA Section
			e) 프로세싱 지시자
			f) 주석
			g) 공백 문자열

	(4) DTD 
		- 종류
			내부 DTD
			외부 DTD
				
		- 구성 요소
			엘리먼트 선언
			속성(attribute) 선언
			Entity 선언
			Notation 선언
			프로세싱 지시자
			파라미터 엔티티 참조
			주석
			공백
			conditional section
				
		- 문서 유형 선언
			<!DOCTYPE 루트엘리먼트명 SYSTEM 또는 PUBLIC "식별자">
			DOCTYPE은 반드시 대문자
			SYSTEM은 특정 단체나 업체내부에서 사용되는 경우
			PUBLIC은 공개
			식별자는 다운로드 받을 수 있는 경로
			PUBLIC일 경우 추가된 식별자
			+-//DTD를 개발 및 유지보수 업체명//DTD명 및 버전번호//사용된 언어

		- 엘리먼트 선언
			<!ELEMENT 엘리먼트명 컨텐트 유형>
			컨텐트 유형
				#PCDATA	: 내용으로 문자데이터만 갖는 엘리먼트
				자식 엘리먼트
					, : 작성 순서
						| : 선택
						? : 생략하거나 한번만 작성
						+ : 한번 이상
						* : 생략하거나 여러번 작성
						기호 없음 : 단 한번만 작성
					EMPTY
					MIXED : 문자데이터 또는 자식엘리먼트를 혼합형태
					ANY

			- ATTRIBUTE 선언
				<!ATTLIST 엘리먼트명 속성명 속성유형 디폴트선언>
				속성 유형
					CDATA	: 문자 데이터
					ENUMERATION	: dtd에 나열된 값 중 하나가 와야 함
					ID : 유일한 값을 지정
					IDREF/IDREFS : ID값을 참조
					NMTOKEN/NMTOKENS : 이름 작성 규칙을 준수하는 
						데이터만 사용
					NOTATION : dtd에 명시적으로 선언된 notation만 사용가능
					ENTITY : dtd에 명시적으로 선언된 entity만 사용 가능
		
		(5) 네임 스페이스
			CML
			-----
			<?xml version=1.0>
			<정보>
				<고유번호>111111-1111111</고유번호>
				<이름>홍길동</이름>
				...
			</정보>
			
			PML
			-----
			<?xml version=1.0>
			<정보>
				<고유번호>LC100</고유번호>
				<이름>캠코더</이름>
				...
			</정보>
			
			OML
			-----
			<?xml version=1.0>
			<주문정보 xmlns:고객="http://www.a.com/2017/Custom"
				xmlns:상품="http://www.a.com/2017/Product"
				xmlns="http://www.a.com/2017/Order">
				<주문번호>1</주문번호>
				<주문수량>10</주문수량>
				<결제>
					<방법>현금</방법>
					<금액>10000000</금액>
				</결제>
				
				<고객:고유번호>111111-1111111</고객:고유번호>
				<고객:이름>홍길동</고객:이름>
				
				<상품:고유번호>LC100</상품:고유번호>
				<상품:이름>캠코더</상품:이름>
				...
			</주문정보>
"""
```

```python
import xml.etree.ElementTree as elemTree
```

```python
# XML을 파싱(Parsing)하는 방법 
#  1. xml파일로 존재하는 경우: parse()
#  2. xml문자열 형태로 존재하는 경우 : fromstring()
```

```python
# xml 태그를 잘 활용하려면 elemTree로 불러와야. parse

tree = elemTree.parse("data/users.xml") 
tree
```

```python
# 메모리에서 생성해서 사용하는 방법 

xmlstr = """<?xml version="1.0" encoding="utf-8" ?>
<users>
	<user grade="gold">
            <name>Kim Cheol Soo</name>
            <age>25</age>
            <birthday>19940215</birthday>
        </user>
        <user grade="diamond">
            <name>Kim Yoo Mee</name>
            <age>21</age>
            <birthday>19980417</birthday>
        </user>
</users>
"""
# 메모리에서 불러올 때 fromstring
tree = elemTree.fromstring(xmlstr)
tree
```

```python
# 태그 검색 및 데이터 다루기  
# user = tree.find("user")
# XPATH : ./ (현재위치), ../ (상위위치)
# user = tree.find("./user[1]") # XML은 1에서 시작
user = tree.find("user[1]") # XML은 1에서 시작
user
dir(user)  # 사용가능 속성 목록
print(user.tag)
print(user.attrib)
print(user.get("grade"))  # 키의 값 가져오기. .get 

username = user.find("name")
print(username.text)

userage = user.find("age")
print(userage.text)
```

```python
# 태그의 조건으로 검색 
user = tree.find("user[2]")   # 태그의 인덱스 

user = tree.find("user[@grade]")  # @속성에 바로 접근
print(user.attrib)

user = tree.find("user[@grade][2]")  # @속성에 바로 접근
print(user.attrib)
```

```python
# 여러 개의 태그를 한꺼번에 가져오기 

users = tree.findall("./user")
users

for user in users:
    print(user.attrib)
    print(user.find("name").text)
```

```python
# attribute 값 읽기 
tree = elemTree.parse("data/users.xml") 
users = tree.getroot() # 상위 
print(users)
print(users.tag)
print(users.attrib) # users는 가장 상위라 속성 없음. 

user = users.find("./user")
print(user.get("grade")) # 첫 유저의 속성에 접근
print(user.keys()) # 키를 모두 보여주기 
print(user.items()) # 값을 모두 보여주기
```

```python
j1 = {"name": "홍길동", "birth":"0524", "age":20}
print(type(j1))
print(j1)

j2 = json.dumps(j1)
print(type(j2))  # 문자열 
print(j2)
```

```python
json.dumps([1,2,3])  # 리스트가 문자열로 묶임
json.dumps((4,5,6))  # 튜플로 문자열로 묶임

```

```python
# 영화 리뷰 
# https://movie.daum.net/moviedb/grade?movieId=94484&type=netizen&page=1
url ="https://movie.daum.net/moviedb/grade?movieId=94484&type=netizen&page=1"

review_list =[] # 리뷰를 모으기 위한 리스트 

for n in range(1, 6):  # 5회 반복 
    target = urlopen("https://movie.daum.net/moviedb/grade?movieId=94484&type=netizen&page=[]".format(n)) 
    soup = BeautifulSoup(target, "html.parser")
    
    reviews = soup.select("p.desc_review")
    for review in reviews:
        str = review.get_text().strip()  # .strip() 은 공백 제거에 사용
#             print(str)
        review_list.append(str)

f = open("data/아쿠아맨.txt", "w")
for review in review_list:
         f.write(review + "\n")
f.close()
```

```python
# 2017년 5월 1일부터 100간의 데이터 수집 
date = pd.date_range("2017-5-1", periods=100, freq="D")
date
```

```python
from tqdm import tqdm_notebook   # 기본 모듈, 현재 진행 상황을 그래프로 보여줌
import urllib 

movie_date = []
movie_name = []
movie_point = []

for today in tqdm_notebook(date): # 여기서 today는 시계열 날짜 타빙으로 되어 있음. date type
    url = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date={dete}"
    page = urlopen(url.format(dete=urllib.parse.quote(today.strftime("%Y%m%d")))) 
    # quote 홑따옴표 붙이는 함수. 문자열로 변환 필요. strftime (datetime 에서from, str로 변환) cf. strptime()
    soup = BeautifulSoup(page, "html.parser")
    
    cnt = len(soup.find_all("td", "point")) # 날짜마다 페이지 내의 영화 수가 다 다르므로 가각 구해줘야.
    
    movie_date.extend([today for n in range(0, cnt)])
    movie_name.extend([soup.find_all("div", "tit5")[n].a.string for n in range(0, cnt)])
    movie_point.extend([soup.find_all("td", "point")[n].string for n in range(0, cnt)])
    
```

```python
len(movie_date), len(movie_name), len(movie_point)
```

```python
movie = pd.DataFrame({"date":movie_date, "name":movie_name, "point":movie_point})
movie.head()
movie.info()

# point는 객체형(object)이면 계산이 안 됨. 숫자형으로 변환해줘야 함. 
```

```python
movie["point"] = movie["point"].astype(float)  # point객체를 float 실수형으로 변환
movie.info()
```

```python
# 영화별로 점수의 합계를 보고 싶다. (합계가 높은 순 또는 낮은 순으로)
movie.head()

m_point = pd.pivot_table(movie, index=["name"], aggfunc = np.mean)
m_point.head()

m_best = m_point.sort_values(by="point", ascending=True)
m_best.head(10)
```
